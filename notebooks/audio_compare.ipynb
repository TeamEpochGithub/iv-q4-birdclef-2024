```python
import numpy as np
import librosa
from pathlib import Path

file_list = [file.stem for file in Path("../data/raw/2024/unlabeled_soundscapes").glob("*.ogg")]
#print(file_list)

test_soundscapes = [np.abs(librosa.load(f"../data/raw/2024/unlabeled_soundscapes/{file_list[i]}.ogg", sr=32000)[0]) for i in range(len(file_list[:1000]))]


#Calculate avg and std of the test soundscapes
test_soundscapes_avg = [np.mean(test_soundscapes[i]) for i in range(len(test_soundscapes))]
test_soundscapes_std = [np.std(test_soundscapes[i]) for i in range(len(test_soundscapes))]

# Print them
print(test_soundscapes_avg)
print(test_soundscapes_std)

#Take another mean again
test_soundscapes_avg = np.mean(test_soundscapes_avg)
test_soundscapes_std = np.mean(test_soundscapes_std)

print(test_soundscapes_avg)
print(test_soundscapes_std)


```

```python
# Load the 

train_audio = Path("../data/raw/2024/train_audio")
#Get all folders in train audio

train_audio_folders = [folder for folder in train_audio.glob("*")]

total_mean = []
total_std = []

for folder in train_audio_folders:
    file_list2 = [file.stem for file in Path(folder).glob("*.ogg")]
    #print(file_list2)
    test = [np.abs(librosa.load(f"{folder}/{file_list2[i]}.ogg", sr=32000)[0]) for i in range(len(file_list2[:15]))]
    
    curr_mean = [np.mean(test[i]) for i in range(len(test))]
    curr_std = [np.std(test[i]) for i in range(len(test))]
    #print(curr_mean)
    
    #Take another mean
    curr_mean = np.mean(curr_mean)
    curr_std = np.mean(curr_std)
    
    total_mean.append(curr_mean)
    total_std.append(curr_std)
    
    
result_mean = np.mean(total_mean)
result_std = np.mean(total_std)

print(result_mean)
print(result_std)
    
    
    



# #Calculate avg and std of the test asbfly
# test_asbfly_avg = [np.mean(test_asbfly[i][0]) for i in range(len(test_asbfly))]
# test_asbfly_std = [np.std(test_asbfly[i][0]) for i in range(len(test_asbfly))]
# 
# # Print them
# print(test_asbfly_avg)
# print(test_asbfly_std)
# 
# #Take another mean again
# test_asbfly_avg = np.mean(test_asbfly_avg)
# test_asbfly_std = np.mean(test_asbfly_std)
# 
# print(test_asbfly_avg)
# print(test_asbfly_std)

```
