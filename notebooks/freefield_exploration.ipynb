```python
import os 
import pandas as pd
import numpy as np

freefield_audio = "../data/raw/freefield/train_audio"
freefield_metapath = "../data/raw/freefield/train_metadata.csv"
meta_df = pd.read_csv(freefield_metapath)
```

```python
#Get all the audio files in the freefield dataset with .wav extension
freefield_audio_files = [f for f in os.listdir(freefield_audio) if f.endswith('.wav')]
freefield_meta_files = [f for f in os.listdir(freefield_audio) if f.endswith('.json')]
print(f"Number of audio files: {len(freefield_audio_files)}")
print(f"Number of metadata files: {len(freefield_meta_files)}")
```

```python
#Get the metadata for the audio files
meta_df.info()
meta_df["metaid"] = meta_df["itemid"].astype(str) + ".json"
meta_df["filename"] = meta_df["itemid"].astype(str) + ".wav"
```

```python
# sort by item id
meta_df["rating"] = 5
meta_df = meta_df.sort_values(by="itemid").reset_index(drop=True)
meta_df.head()
```

```python
#Save meta_Df to csv
meta_df.to_csv(freefield_metapath, index=False)
```

```python
#Setup metadata

#Get the metadata for the audio files

#Load the json files and extract the metadata
import json
import os

meta_dict = {}

for file in freefield_meta_files:
    with open(os.path.join(freefield_audio, file), 'r') as f:
        data = json.load(f)
        meta_dict[file] = data
```
