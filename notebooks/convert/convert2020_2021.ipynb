```python
import pandas as pd

meta = pd.read_csv('../data/raw/2020/train.csv')
meta.head()
```

```python
source_path = '../data/raw/2020/train_audio_44/'
import glob
target_path = '../data/raw/2020/train_audio'
from tqdm.notebook import tqdm
import numpy as np
import librosa
import soundfile
import os

for file in tqdm(glob.glob(source_path+'*/*.mp3')):
    audio, _ = librosa.load(file,duration=5,sr=44100)
    audio = librosa.resample(audio, orig_sr=44100, target_sr=32000).astype(np.float32)
    folder = (target_path+'/'+'/'.join(file.split('/')[-2:-1]))
    tgt_name = (target_path+'/'+'/'.join(file.split('/')[-2:])[:-3]+'ogg')
    if not os.path.exists(folder):
        os.makedirs(folder)
    soundfile.write(tgt_name,audio,32000)
```
